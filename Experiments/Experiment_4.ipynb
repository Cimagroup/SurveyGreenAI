We describe a synthetic dataset

```python
X, y = make_classification(
    n_samples=1000, 
    n_features=2,
    n_informative=2,
    n_redundant=0,
    n_clusters_per_class=2,
    weights=[0.75, 0.25],
    flip_y=0,
    random_state=423)
plot_dataset(X[y==0],X[y==1])
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=2024)
```

We set the arguments for the reduction

For each reduction method, the experiment follows the following steps:

1. Reduce the training dataset with that method.
2. Save the time needed to reduce the dataset.
3. Plots the full and reduced dataset.
4. Train the neural network over the reduced dataset.
5. Save the time needed to train the network.
6. Save the main performance metrics evaluated over the test set (accuracy, sensitivity, specificity and F1-score)
7. Repeat the steps 1-6 a total of n_iter times.
8. Save the mean and variance of each measure in an excel file called "metrics.xlsx".

The experiment depends on a set of parameter, that can be changed in the "exp_functions.py" file. For example, you can set how many reductions are done for each method by changing the n_iter argument, or you can decide the size of the reduced dataset by changing the perc (percentage) parameter.
